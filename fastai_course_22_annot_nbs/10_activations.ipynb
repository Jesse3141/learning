```python
def conv(ni, nf, ks=3, act=True):
    res = nn.Conv2d(ni, nf, stride=2, kernel_size=ks, padding=ks//2)
    if act: res = nn.Sequential(res, nn.ReLU())
    return res

def cnn_layers():
    return [
        conv(1 ,8, ks=5),        #14x14
        conv(8 ,16),             #7x7
        conv(16,32),             #4x4
        conv(32,64),             #2x2
        conv(64,10, act=False),  #1x1
        nn.Flatten()]
```

## Hooks

### Manual insertion

```python
class SequentialModel(nn.Module):
    def __init__(self, *layers):
        super().__init__()
        self.layers = nn.ModuleList(layers)
        self.act_means = [[] for _ in layers]
        self.act_stds  = [[] for _ in layers]
        
    def __call__(self, x):
        for i,l in enumerate(self.layers):
            x = l(x)
            self.act_means[i].append(to_cpu(x).mean())
            self.act_stds [i].append(to_cpu(x).std ())
        return x
    
    def __iter__(self): return iter(self.layers)
```

```python
set_seed(1)
model = SequentialModel(*cnn_layers())
learn = fit(model)
```

### Pytorch hooks

Hooks are PyTorch object you can add to any nn.Module. A hook will be called when a layer, it is registered to, is executed during the forward pass (forward hook) or the backward pass (backward hook). Hooks don't require us to rewrite the model.

```python
set_seed(1)
model = nn.Sequential(*cnn_layers())
```

A hook is attached to a layer, and needs to have a function that takes three arguments: module, input, output. Here we store the mean and std of the output in the correct position of our list.

```python
act_means = [[] for _ in model]
act_stds  = [[] for _ in model]
```

```python
act_means
```

```python
def append_stats(i, mod, inp, outp):
    act_means[i].append(to_cpu(outp).mean())
    act_stds [i].append(to_cpu(outp).std())
```

```python
for i,m in enumerate(model): m.register_forward_hook(partial(append_stats, i))
```

```python
fit(model)
```

```python
for o in act_means: plt.plot(o)
plt.legend(range(5));
```

### Hook class

We can refactor this in a Hook class. It's very important to remove the hooks when they are deleted, otherwise there will be references kept and the memory won't be properly released when your model is deleted.

```python
#| export
class Hook():
    ''' a more flexible hook that can be used to store info inside the hook instance,
    rather than just within the module. 
    self.hook stores the handle returned by registering.
    the hook registered with whatever hook func was passed, but with 
    hook = self.
    a method to remove and overwrite del. unregisters the hook from the module.'''
    def __init__(self, m, f): self.hook = m.register_forward_hook(partial(f, self))
    def remove(self): self.hook.remove()
    def __del__(self): self.remove()
```

```python
def append_stats(hook, mod, inp, outp):
    if not hasattr(hook,'stats'): hook.stats = ([],[])
    acts = to_cpu(outp)
    hook.stats[0].append(acts.mean())
    hook.stats[1].append(acts.std())
```

```python
set_seed(1)
model = nn.Sequential(*cnn_layers())
```

```python
hooks = [Hook(l, append_stats) for l in model[:5].children()]
```

```python
learn = fit(model)
```

```python
for h in hooks:
    plt.plot(h.stats[0])
    h.remove()
plt.legend(range(5));
```

```python
#| export
class Hooks(list):
    '''our list of hooks but also a context manager.
    init with module and function'''
    def __init__(self, ms, f): super().__init__([Hook(m, f) for m in ms])
    def __enter__(self, *args): return self
    def __exit__ (self, *args): self.remove()
    def __del__(self): self.remove()
    def __delitem__(self, i):
        self[i].remove()
        super().__delitem__(i)
    def remove(self):
        for h in self: h.remove()
```

```python
set_seed(1)
model = nn.Sequential(*cnn_layers())
```

```python
with Hooks(model, append_stats) as hooks:
    fit(model)
    fig,axs = plt.subplots(1,2, figsize=(10,4))
    for h in hooks:
        for i in 0,1: axs[i].plot(h.stats[i])
    plt.legend(range(6));
```

### HooksCallback

```python
#| export
class HooksCallback(Callback):
    '''created hook as a callback.
    filter: the layers to which to apply: a condition such as fc.risinstance(nn,.Conv2d)'''
    def __init__(self, hookfunc, mod_filter=fc.noop, on_train=True, on_valid=False, mods=None):
        fc.store_attr()
        super().__init__()
    
    def before_fit(self, learn):
        if self.mods: mods=self.mods
        else: mods = fc.filter_ex(learn.model.modules(), self.mod_filter)
        self.hooks = Hooks(mods, partial(self._hookfunc, learn))

    def _hookfunc(self, learn, *args, **kwargs):
        if (self.on_train and learn.training) or (self.on_valid and not learn.training): self.hookfunc(*args, **kwargs)

    def after_fit(self, learn): self.hooks.remove()
    def __iter__(self): return iter(self.hooks)
    def __len__(self): return len(self.hooks)
```

```python
hc = HooksCallback(append_stats, mod_filter=fc.risinstance(nn.Conv2d))
```

```python
set_seed(1)
model = nn.Sequential(*cnn_layers())
fit(model, xtra_cbs=[hc]);
```

## Histograms

```python
#| export
def append_stats(hook, mod, inp, outp):
    '''our standard stats hook. 
    stats[2] hold a list of histogram counts of a hist of 40 bins between 0 and 10'''
    if not hasattr(hook,'stats'): hook.stats = ([],[],[])
    acts = to_cpu(outp)
    hook.stats[0].append(acts.mean())
    hook.stats[1].append(acts.std())
    hook.stats[2].append(acts.abs().histc(40,0,10))
```

```python
set_seed(1)
model = nn.Sequential(*cnn_layers())
hc = HooksCallback(append_stats, mod_filter=fc.risinstance(nn.Conv2d))
fit(model, xtra_cbs=[hc]);
```

```python
#| export
# Thanks to @ste for initial version of histgram plotting code
def get_hist(h):
    ''' each column represents the histogram of activations for a batch,
    and each row represents a specific bin across all batches.'''
    return torch.stack(h.stats[2]).t().float().log1p()
```

### nb 10 things to remember:

**good training**
>the avg activation should be about 0, and the std should be 1. if all 0 - it means nothing is happening. 

**callbacks**
>common: cbs = [TrainCB(), DeviceCB(), metrics, ProgressCB(plot=True)]
>
**hooks**
> essentially a type of call back.
> A hook is **attached to a layer**, and needs to have a function that takes three arguments: module, input, output.
> pytorch automatically calls all hooks, with mod, inp, outp as args.
> >in our case, we are goping to store info about the activations so as to see what's happening in the model.
> any additional args for the func have to inserted differently, for instance creating a func instance with partial.
> 
> in pytorch, adding a hook is called registering (forward or back), and executed when the layer it is attached to is called.
> note: the register method returns a 'handle' (similar to apointer) that must be stored if you want to be able to remove the hook later.
> we made hooks an object owned by a module.
> then can define functions which alter the hook object, for instance, saving stats. that's why we need the function that ends up in the hook to have itself as a pre-defined param.
> 
> and therefore, we will need to keep a list of pointers to the hooks so we can delete them later and free up the memory: for that we create the **hooks class** which can handle registering and removing.
>
> our **get_hist** creates a histogram of the activations of all batches. **get_min** simply tells us what proportion of actiavations were in the two lowest categories.

## ActivationStats

```python
#|export
class ActivationStats(HooksCallback):
    '''specifically a call back of activations hook'''
    def __init__(self, mod_filter=fc.noop): super().__init__(append_stats, mod_filter)

    def color_dim(self, figsize=(11,5)):
        fig,axes = get_grid(len(self), figsize=figsize)
        for ax,h in zip(axes.flat, self):
            show_image(get_hist(h), ax, origin='lower')

    def dead_chart(self, figsize=(11,5)):
        fig,axes = get_grid(len(self), figsize=figsize)
        for ax,h in zip(axes.flatten(), self):
            ax.plot(get_min(h))
            ax.set_ylim(0,1)

    def plot_stats(self, figsize=(10,4)):
        fig,axs = plt.subplots(1,2, figsize=figsize)
        for h in self:
            for i in 0,1: axs[i].plot(h.stats[i])
        axs[0].set_title('Means')
        axs[1].set_title('Stdevs')
        plt.legend(fc.L.range(self))
```

```python
astats = ActivationStats(fc.risinstance(nn.Conv2d))
```

```python
set_seed(1)
model = nn.Sequential(*cnn_layers())
fit(model, xtra_cbs=[astats]);
```
